```
"""
Array
"""

class Array:

    def __init__(self, tam):
        self.__datos = [0 for x in range(tam)] 

    def get_item(self, index):
        dato = 0
        if index >= 0 and index < len(self.__datos):
            dato = self.__datos[index]
        else:
            print("Eror en el indice")
        return dato

    def set_item(self, dato, index):
        if index >= 0 and index < len(self.__datos):
            self.__datos[index] = dato
        else:
            print("Eror en el indice")

    def get_Length(self):
        return len(self.__datos)

    def clear(self, dato):
        self.__datos = [dato for x in range(tam)]  

    def __iter__(self):
        return _getIterator(self.__datos)

class _getIterator:
    def __init__(self, arreglo):
        self.__arreglo = arreglo
        self.__indice_actual = 0

    def __iter__(self):
        return self

    def __next__(self):
        if self.__indice_actual < len(self.__arreglo):
            dato = self.__arreglo[self.__indice_actual]
            self.__indice_actual += 1
            return dato
        else:
            raise StopIteration

    def getIterator(self):
        return _getIterator(self.__datos)

"""
mi_arreglo = Array(10)
print(mi_arreglo.get_item( 20 ))
mi_arreglo.set_item( 2345 , 3 )
print(mi_arreglo.get_item( 3 ))

print("--------")
for x in mi_arreglo:
    print(x)
"""
```

```
class Empleado:
    def __init__( self , num , nom , pat , mat , he , sb , anio ):
        self.__num_trabajador = num
        self.__nombres = nom
        self.__paterno = pat
        self.__materno = mat
        self.__horas_extra = he
        self.__sueldo_base = sb
        self.__anio_ingreso = anio
    # Metodos get y set
    def getNum_trabajador ( self ):
        return self.__num_trabajador
    def setNum_trabajador ( self, num ):
        self.__num_trabajador = num

    def getNombre( self ):
        return self.__nombres

    def setNombre( self, nom ):
        self.__nombres = nom

    def getPaterno(self):
        return self.__paterno

    def setPaterno( self, pat ):
        self.__paterno = pat

    def getMaterno( self ):
        return self.__materno

    def setMaterno( self, mat ):
        self.__materno = mat

    def getHextra( self ):
        return self.__horas_extra

    def setHextra( self, he ):
        self.__horas_extra = he

    def getSalariobase( self ):
        return self.__sueldo_base

    def setSalariobase( self, sb ):
        self.__sueldo_base = sb

    def getAnioingreso( self ):
        return self.__anio_ingreso

    def setAnioingreso(self, anio):
        self.__anio_ingreso = anio
    # to_string
    def to_string(self):
        return f"{self.__num_trabajador}  {self.__nombres} {self.__paterno} {self.__materno} " \
               f"{self.__horas_extra} {self.__sueldo_base} {self.__anio_ingreso}"

    # Metodos de uso general: Ca
    def sueldo_mensual( self ):
        antiguedad = ( (2020 - self.__anio_ingreso) * 0.3 ) * self.__sueldo_base
        hextra = self.__horas_extra * 276.5
        return self.__sueldo_base + hextra + antiguedad
```

```
class Sueldos:
    def __init__(self):
        self.__data = None
    
    def cargar_datos( self , ruta_archivo):
        archivo = open(ruta_archivo, "rt", encoding = 'latin1')
        lineas = archivo.readlines()
        archivo.close()
        self.__data = Array(len(lineas) - 1)
        #La tarea
        for iter in range(1,len(lineas)):
            lineas[iter] = lineas[iter].split(",")

            for posicion in [-1,-2,-3]:#SL
                lineas[iter][posicion] = int(lineas[iter][posicion])
            colaboradorsb = Empleado(lineas[iter][0],lineas[iter][1],lineas[iter][2],
                                 lineas[iter][3],lineas[iter][4],lineas[iter][5],lineas[iter][6])

            self.__data.set_item(colaboradorsb,iter-1)
    
    def calcular_sueldos ( self ):

        print("SUELDOS NETOS DE EMPLEADOS: ")
        for emp in self.__data:
            print(f"id de empleado: {emp.getNum_trabajador()}, Nombre del empleado: {emp.getNombre()}, Apellidos {emp.getPaterno()} {emp.getMaterno()} , SUELDO NETO DEL EMPLEADO = ${emp.sueldo_mensual()}")
    
    def reportar_mayor_ant( self ):
        mayor = 0
        for emp in self.__data:
            antiguedad = 2020 - emp.getAnioingreso()
            if antiguedad > mayor:
                mayor = antiguedad
        print("Empledo con mayor antiguedad: \n ")
       
        for emp in self.__data:
            antiguedad = 2020 - emp.getAnioingreso()
            if antiguedad == mayor:
                print(f"id de empleado: {emp.getNum_trabajador()}, Nombre {emp.getNombre()} Apellido Paterno: {emp.getPaterno()}, Apellido Materno: {emp.getMaterno()}, Tiene una antiguedad de: {mayor} años de antiguedad \n ○2")
    
    def reportar_menor_ant( self ):
        menor = 100
        for emp in self.__data:
            antiguedad = 2020 - emp.getAnioingreso()

            if antiguedad < menor:
                menor = antiguedad
        print("Empledo con menor antiguedad: \n ")
        for emp in self.__data:
            antiguedad = 2020 - emp.getAnioingreso()

            if antiguedad == menor:
                print(f"id de empleado: {emp.getNum_trabajador()} , Nombre {emp.getNombre()}, Apellido Paterno: {emp.getPaterno()} , Apellido Materno: {emp.getMaterno()}  Tienen: {menor} años de antiguedad \n ")

ejemplo = Sueldos()
ejemplo.cargar_datos('junio.dat')
ejemplo.calcular_sueldos()
ejemplo.reportar_mayor_ant()
ejemplo.reportar_menor_ant()
```
